```python
import csv
import numpy as np
import matplotlib.pyplot as plt

# Import collected data
file = "Values.csv"
data = np.genfromtxt(file,delimiter=',')
tdata = data[:,4]
ydata = data[:,5]
fig, (ax1) = plt.subplots(1, 1)
# #c0c0c0
ax1.plot(tdata, ydata, marker=".", color="#c0c0c0", label="Experimental Data")

plt.xlabel("Eps (s)")
plt.ylabel("Kc (°C")
plt.xlim(0, 1)             
plt.legend(title="Temperature (°C)")
plt.show()
```
